```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv',encoding='UTF-8')
exchange_rates.head(5)
```

```python
exchange_rates.tail(5)
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                              'Period\\Unit:': 'Time'},
                     inplace=True)
exchange_rates['Time']=pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
        euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean']=euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.plot(euro_to_dollar['Time'],
        euro_to_dollar['rolling_mean'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],
        euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rm in zip(range(2,7),[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
            euro_to_dollar['US_dollar'].rolling(rm).mean())
    plt.title('Rolling Window:{}'.format(str(rm)), weight='bold')
plt.tight_layout()
plt.show()
```

# Coming Up With an Idea
* We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
* We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2016 and 2009 for comparison. We can use a line plot.
* We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

## Financial Crisis Example

```python
financial_crisis_6_9 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <=
                                  2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2007) & (euro_to_dollar['Time'].dt.year <=2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis_6_9['Time'],
       financial_crisis_6_9['rolling_mean'],
       linewidth=1,color='#A6D785')

ax.plot(financial_crisis_7_8['Time'],
       financial_crisis_7_8['rolling_mean'],
       linewidth=1,color='#e23d28')
ax.axvspan(xmin=733112.0,xmax=733302.0, ymin=0.09,
          alpha=0.3, color='grey')

ax.set_xticklabels([])
ax.set_yticklabels([])

x=732272.0
for year in ['2006','2007','2008','2009','2010']:
    ax.text(x,1.13, year, alpha=0.5, fontsize=11)
    x+=365

y=1.193
for rate in ['1.2','1.3','1.4','1.5']:
    ax.text(732172.0, y, rate, alpha=0.5, fontsize=11)
    
ax.text(732172.0, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
       weight='bold')
ax.text(732172.0, 1.63, "Euro-USD exchange rates between 2006 and 2010",
       size=12)

ax.text(732172.0, 1.07, '@DATAQUEST' + ' '*94 + 'Source: European Central Bank',
       color='#f0f0f0', backgroundcolor='#4d4d4d',
       size=10)

ax.grid(alpha=0.5)
plt.show()
```

```python
bush_omaba_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) &
                                        (euro_to_dollar['Time'].dt.year <= 2021)]

bush = bush_omaba_trump.copy()[bush_omaba_trump['Time'].dt.year <2009
                               ]
obama = bush_omaba_trump.copy()[(bush_omaba_trump['Time'].dt.year >=2009) &
                               (bush_omaba_trump['Time'].dt.year <2017)]
trump = bush_omaba_trump.copy()[(bush_omaba_trump['Time'].dt.year >=2017) &
                               (bush_omaba_trump['Time'].dt.year <2021)
                               ]
```
